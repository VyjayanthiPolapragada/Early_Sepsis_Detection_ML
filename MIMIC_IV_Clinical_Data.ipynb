```python
# ICD-10 code pattern for sepsis (e.g., A41.* = sepsis)
sepsis_icd10 = diagnoses_icd_df[
    diagnoses_icd_df['icd_code'].str.startswith('A41')
]

# Merge with patient IDs
sepsis_patients = sepsis_icd10[['subject_id']].drop_duplicates()
```

```python
# Filter icustays for only sepsis patients
sepsis_icustays = icustays_df[icustays_df['subject_id'].isin(sepsis_patients['subject_id'])]
```

```python
# Join icustay_id with labevents
merged_labs = pd.merge(labevents_df, sepsis_icustays, on='subject_id')

# Keep only lactate entries (use d_labitems to find itemid)
lactate_ids = d_labitems_df[
    d_labitems_df['label'].str.contains('lactate', case=False, na=False)
]['itemid'].tolist()

# Filter lactate measurements in first 6 hours of ICU
merged_labs['charttime'] = pd.to_datetime(merged_labs['charttime'])
merged_labs['intime'] = pd.to_datetime(merged_labs['intime'])

lactate_6hr = merged_labs[
    (merged_labs['itemid'].isin(lactate_ids)) &
    ((merged_labs['charttime'] - merged_labs['intime']).dt.total_seconds() <= 6 * 3600)
]

# Aggregate features
lactate_features = lactate_6hr.groupby('stay_id')['valuenum'].agg(['mean', 'max', 'min', 'count']).reset_index()
lactate_features.columns = ['stay_id', 'lactate_mean_6h', 'lactate_max_6h', 'lactate_min_6h', 'lactate_count_6h']
```

```python
# Heart rate itemid(s) — find using d_items
hr_ids = d_items_df[d_items_df['label'].str.contains('heart rate', case=False)]['itemid'].tolist()

# Filter and join
hr_events = chartevents_df[
    chartevents_df['itemid'].isin(hr_ids)
]

# Merge with ICU stay times
hr_merged = pd.merge(hr_events, sepsis_icustays[['subject_id', 'stay_id', 'intime']], on='subject_id')

# Convert to datetime
hr_merged['charttime'] = pd.to_datetime(hr_merged['charttime'])
hr_merged['intime'] = pd.to_datetime(hr_merged['intime'])

# First 6 hours of ICU
hr_6hr = hr_merged[(hr_merged['charttime'] - hr_merged['intime']).dt.total_seconds() <= 6 * 3600]

# Aggregate
hr_features = hr_6hr.groupby('stay_id_x')['valuenum'].agg(['mean', 'max', 'min', 'count']).reset_index()
hr_features.columns = ['stay_id_x', 'hr_mean_6h', 'hr_max_6h', 'hr_min_6h', 'hr_count_6h']
```

```python
sepsis_features = pd.merge(
    lactate_features, 
    hr_features, 
    left_on='stay_id', 
    right_on='stay_id_x', 
    how='outer'
)
```
